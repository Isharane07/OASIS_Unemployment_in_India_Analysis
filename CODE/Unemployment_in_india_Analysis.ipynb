```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8')
```

```python
df = pd.read_csv('Unemployment in India.csv')
df.head()
```

```python
df.columns = df.columns.str.strip()
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
df.dtypes
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(7,4))
sns.lineplot(
    x='Date',
    y='Estimated Unemployment Rate (%)',
    data=df
)
plt.title("Unemployment Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
covid_df = df[df['Date'] >= '2020-03']
```

```python
plt.figure(figsize=(7,4))
sns.lineplot(
    x='Date',
    y='Estimated Unemployment Rate (%)',
    data=covid_df,
    color='red'
)
plt.title("Unemployment Rate During Covid-19")
plt.show()
```

```python
state_avg = df.groupby('Region')['Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False)

plt.figure(figsize=(6,5))
state_avg.head(10).plot(kind='bar')
plt.title("Top 10 States with Highest Unemployment")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
df_sorted = df.sort_values('Date')

df_sorted['Rolling_Avg'] = (
    df_sorted['Estimated Unemployment Rate (%)']
    .rolling(window=6)
    .mean()
)
```

```python
plt.figure(figsize=(7,4))
plt.plot(df_sorted['Date'], df_sorted['Estimated Unemployment Rate (%)'], alpha=0.3, label='Actual')
plt.plot(df_sorted['Date'], df_sorted['Rolling_Avg'], color='red', label='6-Month Rolling Avg')
plt.title("Unemployment Rate Trend (Smoothed)")
plt.legend()
plt.show()
```

```python
pre_covid = df[df['Date'] < '2020-03']
covid = df[(df['Date'] >= '2020-03') & (df['Date'] <= '2021-06')]
post_covid = df[df['Date'] > '2021-06']

print("Pre-Covid Avg:", pre_covid['Estimated Unemployment Rate (%)'].mean())
print("Covid Avg:", covid['Estimated Unemployment Rate (%)'].mean())
print("Post-Covid Avg:", post_covid['Estimated Unemployment Rate (%)'].mean())
```

```python
df['Period'] = 'Pre-Covid'

df.loc[(df['Date'] >= '2020-03') & (df['Date'] <= '2021-06'), 'Period'] = 'Covid'

df.loc[df['Date'] > '2021-06', 'Period'] = 'Post-Covid'
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(
    x='Period',
    y='Estimated Unemployment Rate (%)',
    data=df,
    showfliers=False   #  outliers hide
)
plt.title("Unemployment Rate Distribution Across Periods")
plt.xlabel("Period")
plt.ylabel("Unemployment Rate (%)")
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(6,4))
sns.violinplot(
    x='Period',
    y='Estimated Unemployment Rate (%)',
    data=df,
    inner='quartile'
)
plt.title("Unemployment Rate Density by Period")
plt.tight_layout()
plt.show()
```

```python
state_time = df.pivot_table(
    values='Estimated Unemployment Rate (%)',
    index='Region',
    columns=df['Date'].dt.year,
    aggfunc='mean'
)

plt.figure(figsize=(10,6))
sns.heatmap(state_time, cmap='Reds')
plt.title("State-wise Unemployment Heatmap (Year-wise)")
plt.show()
```

```python
corr = df[['Estimated Unemployment Rate (%)', 'Estimated Employed']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Analysis")
plt.show()
```
